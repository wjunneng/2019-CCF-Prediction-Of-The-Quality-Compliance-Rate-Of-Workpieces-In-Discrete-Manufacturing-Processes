```python
import pandas as pd
pd.set_option('display.max_column',30)
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pylab import rcParams ##matplotlib
rcParams['figure.figsize'] = 12, 8
```

```python
project_path = '/home/wjunneng/Ubuntu/2019-CCF-Prediction-Of-The-Quality-Compliance-Rate-Of-Workpieces-In-Discrete-Manufacturing-Processes'

first_round_testing_data_path = project_path + '/data/original/first_round_testing_data.csv'
first_round_training_data_path = project_path + '/data/original/first_round_training_data.csv'

first_round_testing_data = pd.read_csv(first_round_testing_data_path)
first_round_training_data = pd.read_csv(first_round_training_data_path)

train = pd.read_csv(first_round_training_data_path)[['Parameter'+str(i) for i in range(1,11)]+['Quality_label']]
test = pd.read_csv(first_round_testing_data_path)
```

```python
train.head()
```

## Label Distribution

```python
train['Quality_label'].value_counts().plot(kind='bar')
```

```python
train['label'] = train['Quality_label'].map({
    'Excellent':1,
    'Good':2,
    'Pass':3,
    'Fail':4
})
```

```python
train.describe()
```

## `Correlation`

```python
plt.figure(figsize=(12,8))
correlation = train.corr()
mask = np.zeros_like(correlation)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data=correlation.round(3),mask=mask,vmin=-0.1,vmax=0.1,annot=True, linewidths=.1)
```

## 判断是否是类别变量

```python
train.nunique().plot(kind='bar')
```

## `Parameter` 与 `label`

```python
for i in ['Parameter'+str(i) for i in range(1,11)]:
    sns.scatterplot(x='Quality_label', y=i, data=train)
    plt.title(i)
    plt.show()
```

## 训练集和测试集分布比较

```python
for i in ['Parameter'+str(i) for i in range(5,11)]:
    train[i].value_counts().plot(kind='bar',alpha=0.5, label='Train')
    test[i].value_counts().plot(kind='bar',alpha=0.5,color='r', label='Test')
    plt.title(i)
    plt.legend()
    plt.show()
```

```python
for i in ['Parameter'+str(i) for i in range(1,11)]:
    sns.kdeplot(train[i], label='Train',alpha=0.8)
    sns.kdeplot(test[i], label='Test',alpha=0.8)
    plt.title(i)
    plt.show()
```

```python
for i in ['Parameter'+str(i) for i in range(1,11)]:
    plt.figure(figsize=(14,8))
    plt.scatter(x=range(len(train)), y=train[i], label='Train')
    plt.scatter(x=range(len(test),len(test)*2), y=test[i], label='Test')
    plt.title(i)
    plt.legend()
    plt.show()
```

```python
for i in ['Parameter'+str(i) for i in range(1,11)]:
    plt.scatter(x=range(len(train)), y=sorted(np.log(train[i])), label='Train')
    plt.scatter(x=range(len(test)), y=sorted(np.log(test[i])), label='Test')
    plt.title(i+'--log')
    plt.legend()
    plt.show()
    
    sm.qqplot(np.log(train[i]), line='s', label='Train')
    plt.legend()
    plt.title(i+'--QQ Plot')
    plt.show()
```

```python
train_label_1 = train[train['label'] == 1]
train_label_2 = train[train['label'] == 2]
train_label_3 = train[train['label'] == 3]
train_label_4 = train[train['label'] == 4] 

for i in ['Parameter'+str(i) for i in range(1,11)]:
    plt.figure(figsize=(14,8))
    plt.scatter(x=range(len(train_label_1)), y=train_label_1[i], label='Train_label_1')
    length = len(train_label_1)
    plt.scatter(x=range(length, length+len(train_label_2)), y=train_label_2[i], label='Train_label_2')
    length = length + len(train_label_2)
    plt.scatter(x=range(length, length+len(train_label_3)), y=train_label_3[i], label='Train_label_3')
    length = length + len(train_label_3)
    plt.scatter(x=range(length, length+len(train_label_4)), y=train_label_4[i], label='Train_label_4')
    length = length + len(train_label_4)
    plt.scatter(x=range(length, length+len(test)), y=test[i], label='Test')
    plt.title(i)
    plt.legend()
    plt.show()
```
